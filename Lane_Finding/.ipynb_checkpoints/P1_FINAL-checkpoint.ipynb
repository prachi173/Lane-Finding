```python
#Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import sys
%matplotlib inline
import os
import math
import time
```

<b>FUNCTIONS TO PROCESS THE IMAGE</b><br><br>
<b>Includes the following functions in order of process:</b><br>
1. Convert to grayscale
2. Darken image
3. Gaussian blur
4. Mask image for region of interest
5. Apply CannyEdge detection
6. Draw lines using HoughLines
7. Extrapolate the point on houghlines to create lines

```python
#Functions to process images
#Function that applies gamma correction to an image using look up table
#Not necessary but darkening the video gave better result for this video
def adjust_gamma(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")    
    return cv2.LUT(image, table)

def convert_gray(image):
    #img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    return gray

def gaussian_blur(image):
    kernal_size = 7
    blur_gray = cv2.GaussianBlur(image,(kernal_size, kernal_size), 0 )
    return blur_gray

def canny_edge(image, low_threshold, high_threshold):
    edges = cv2.Canny(image, low_threshold, high_threshold)
    
    return edges

def region_of_interest(image, imshape, vertices):
    mask = np.zeros_like(image)
    if len(imshape) > 2:
        channel_count = imshape[2]
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
            
    #masking the region of interest for Houghlines
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(image, mask)
            
    return masked_edges

def drawing_lines(masked_edges, image):
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 20     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 20 #minimum number of pixels making up a line
    max_line_gap = 300    # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on


    #Drawing lines 
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
    
    return lines, line_image

def extrapolate_lines(lines, line_image, imshape):
    #Averaging & extrapolating to show clear, continuous line through the video
    ymin = imshape[0]
    ymax = imshape[0]
    left_lane = []
    right_lane = []
    left_y = []
    right_y = []
    left_x = []
    right_x = []

    for line in lines:
        for x1,y1,x2,y2 in line:
            lane, intercept = np.polyfit((x1,x2), (y1,y2), 1)
            ymin = min(min(y1,y2), ymin)

            if(lane > 0):
                left_lane += [lane]
                left_y += [y1,y2]
                left_x += [x1,x2]
            else:
                right_lane += [lane]
                right_y += [y1,y2]
                right_x += [x1,x2]

        left_avg = np.mean(left_lane)
        left_y_avg = np.mean(left_y)
        left_x_avg = np.mean(left_x)
        left = left_y_avg - (left_avg * left_x_avg)

        right_avg = np.mean(right_lane)
        right_y_avg = np.mean(right_y)
        right_x_avg = np.mean(right_x)
        right = right_y_avg - (right_avg * right_x_avg)

        if ((len(left_lane) > 0) and (len(right_lane) > 0)):
            upper_left_x = int((ymin - left) / left_avg)
            lower_left_x = int((ymax - left) / left_avg)
            upper_right_x = int((ymin - right) / right_avg)
            lower_right_x = int((ymax - right) / right_avg)

            #Adding extrapolated lanes to line_image
            cv2.line(line_image, (upper_left_x, ymin), (lower_left_x, ymax), color=[0,0,255], thickness = 6)
            cv2.line(line_image, (upper_right_x, ymin), (lower_right_x, ymax), color=[0,0,255], thickness = 6)

    
    return line_image

def process_image(frame, vertices):
    frame1 = np.copy(frame)
    imshape = frame1.shape 
    #Convert to gray and darken it for better result
    gray1 = convert_gray(frame1)
    gray = adjust_gamma(gray1, 0.3)
    #Set the kernal for gaussian blur
    blur_gray = gaussian_blur(gray)
    #canny edge detection
    low_threshold = 70 #Threshold below this is rejected
    high_threshold = 150 #Threshold above this is accepted as edge
    edges = canny_edge(blur_gray,low_threshold,high_threshold)

    #creating a mask
    #Vertices for the region of interest polygon.
    masked_edges = region_of_interest(edges, imshape, vertices)
    lines, line_image = drawing_lines(masked_edges, frame1)

    #Extrapolating lines
    line_image = extrapolate_lines(lines, line_image, imshape)

    #Stack frames and add lines to the original frame
    color_edges = np.dstack((edges,edges,edges))
    draw_lines = cv2.addWeighted(frame1, 0.8, line_image, 1, 0)
    
    return draw_lines
```

First, we will work with the iamges that are given to us in the folder 'test_images'

```python
#First, let's try with the images
image_list = os.listdir('/test_images/')
os.listdir('/test_images/')
```

Let's see the original images and put them in a list

```python
image_list2 = []
for i in image_list:
    img = mpimg.imread("/test_images/"+i)
    image_list2.append(img)
    plt.figure()
    plt.imshow(img)
```

The initial steps are to convert the images to gray and applying gaussian blur. I took an extra step to darken the image for better result.

```python
i = image_list2[3]
gray_img = convert_gray(i)
darken_img = adjust_gamma(gray_img, 0.5)
blur_img = gaussian_blur(darken_img)
plt.figure()
plt.imshow(blur_img, cmap = 'gray')
```

Next, we apply a mask for region of interest.

```python
i = image_list2[3]
vertices_img = np.array([[(130, 540), (450, 320),(505, 320),(855,540)]], dtype=np.int32)
imgshape = i.shape
mask_edges = region_of_interest(blur_img, imgshape, vertices_img)

plt.figure()
plt.imshow(mask_edges, cmap='gray')
```

Now we apply the Canny edge detection to detect edges for our lanes.

```python
i = image_list2[3]
mask_edges = region_of_interest(blur_img, imgshape, vertices_img)
lt = 200
ht = 300
cannyedge = canny_edge(mask_edges, lt, ht)

plt.figure()
plt.imshow(cannyedge, cmap='gray')
```

```python
lines_img, lines_image_img = drawing_lines(cannyedge, i)
for line in lines_img:
    for x1,y1,x2,y2 in line:
        cv2.line(lines_image_img, (x1, y1), (x2, y2), color=[255,0,0], thickness=5)
    lines_edges_img = cv2.addWeighted(i, 0.8, lines_image_img, 1, 0)
    plt.imshow(lines_edges_img)
```

The next steps are to plot houghlines on the images and extrapolate the points to draw lines on the image.

```python
i = 0
for i in image_list2:
    img_draw_lines = process_image(i, vertices_img)
    plt.figure()
    plt.imshow(img_draw_lines)
```
